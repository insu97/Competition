```python
import random
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


plt.rc('font', family='Malgun Gothic')  # 폰트 지정
plt.rc('axes', unicode_minus=False)  # 마이너스 폰트 설정
%config InlineBackend.figure_format = 'retina'  # 그래프 글씨 뚜렷


def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)


seed_everything(42)

df_train = pd.read_csv("../data/preprocessed/df_train001.csv")
df = pd.read_csv("../data/preprocessed/df.csv")
```

---

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X = df_train.drop(columns = ['일시','평균기온'], axis=1)
y = df_train['평균기온']
```

```python
# 그라디언트 부스팅 회귀 모델 생성 및 학습
gb_regressor = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gb_regressor.fit(X, y)
```

```python
y_pred = gb_regressor.predict(df.drop(columns = ['일시'], axis=1))
```

```python
plt.plot(y_pred)
```

---

```python
sub_file = pd.read_csv("../data/original/sample_submission.csv")
```

```python
sub_file['평균기온'] = y_pred
```

```python
sub_file.to_csv("../data/sub/016_df_gradientboosting_001.csv",
                index=False, encoding='utf-8')
```
