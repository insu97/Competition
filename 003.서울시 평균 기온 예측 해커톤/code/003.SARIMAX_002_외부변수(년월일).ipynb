```python
import random
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')


plt.rc('font', family='Malgun Gothic')  # 폰트 지정
plt.rc('axes', unicode_minus=False)  # 마이너스 폰트 설정
%config InlineBackend.figure_format = 'retina'  # 그래프 글씨 뚜렷


def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)


seed_everything(42)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df_train = pd.read_csv("../data/preprocessed/df_train001.csv")
```

---

```python
df_train['년'] = df_train['일시'].str.split('-', expand=True)[0]
df_train['월'] = df_train['일시'].str.split('-', expand=True)[1]
df_train['일'] = df_train['일시'].str.split('-', expand=True)[2]
```

```python
df_train[['년', '월', '일']] = df_train[['년', '월', '일']].astype('float')
```

```python
df_train.head()
```

```python
df_train['일시'] = pd.to_datetime(df_train['일시'])
df_train.set_index('일시', inplace=True)
# df_train = df_train.astype('float')
```

```python
train_size = int(len(df_train) * 0.8)
y_train, y_test = df_train['평균기온'][:train_size], df_train['평균기온'][train_size:]
```

```python
# order: (p, d, q), seasonal_order: (P, D, Q, S)
order = (1, 0, 1)
seasonal_order = (1, 0, 1, 12)  # 주간 계절성을 고려하기 위해 S=7
exog_vars_train = df_train[['년', '월', '일']][:train_size]
exog_vars_test = df_train[['년', '월', '일']][train_size:]
# exog_vars_train.reset_index(drop=True, inplace=True)
# exog_vars_test.reset_index(drop=True, inplace=True)
```

```python
model = SARIMAX(y_train, order=order,
                seasonal_order=seasonal_order, exog=exog_vars_train)
results = model.fit()
```

```python
pred = results.get_prediction(
    start=y_test.index[0], end=y_test.index[-1], dynamic=False, exog=exog_vars_test)
pred_ci = pred.conf_int()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_train['평균기온'], label='Actual Sales')
plt.plot(pred.predicted_mean, label='Forecast', color='red')
plt.fill_between(pred_ci.index, pred.conf_int()['lower 평균기온'], pred.conf_int()[
                 'upper 평균기온'], color='pink', alpha=0.3)
plt.title('SARIMAX Forecast with External Variable')
plt.legend()
plt.show()
```
