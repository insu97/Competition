```python
import random
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')


plt.rc('font', family='Malgun Gothic')  # 폰트 지정
plt.rc('axes', unicode_minus=False)  # 마이너스 폰트 설정
%config InlineBackend.figure_format = 'retina'  # 그래프 글씨 뚜렷


def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)


seed_everything(42)
```

```python
df_train = pd.read_csv("../data/preprocessed/df_train001.csv")
```

```python
df_train['일시'] = pd.to_datetime(df_train['일시'])
df_train.set_index('일시', inplace=True)
```

```python
df_train.index.freq = 'D'
```

```python
COLORS = {
    'green': '#278F45',
    'red': '#DB4A25', 
    'blue': '#23D3DB',
    'black': '#000000', 
    'yellow': '#DBBE2E',
}
```

# ARIMA 모델
    - AR
        - Autoregressive 모델은 자기 회귀 모델이라고 불린다. 
        - 과거 시점의 자기 자신의 데이터가 현 시점의 자기 자신에게 영향을 미치는 모델
    - MA
        - Moving average 모델은 이동 평균 모델이라고 불리고, 트렌드 즉 추세가 변하는 상황에서 적합한 모델
    - ARMA
        - ARMA는 AR과 MA모델을 합친 모델이다. 
        - 현재 시점의 상태를 파악하는 과거 시점의 자기 자신과 추세까지 전부 반영하겠다는 의미
    - ARIMA
        - 이제까지 AR, MA, ARMA모델의 경우 시계열이 정상성이라는 가정이 있는 상황에서 진행했다면, 
        - ARIMA모델에서는 차분이라는 개념을 통해 non-stationary한 상황에서 좀 더 나은 예측을 하는 것이 목표

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
import itertools
from tqdm import tqdm
```

```python
p = range(0, 4)
d = range(0, 3)
q = range(0, 5)

pdq = list(itertools.product(p,d,q))

aic = []
params = []

with tqdm(total=len(pdq)) as pg:
    for i in pdq:
        pg.update(1)
        try:
            model = ARIMA(df_train['평균기온'], order=(i)).fit()
            aic.append(round(model.aic, 2))
            params.append((i))
        except:
            continue 
```

```python
optimal = [(params[i],j) for i,j in enumerate(aic) if j == min(aic)]
arima = ARIMA(df_train['평균기온'], order = optimal[0][0], freq='D').fit()
arima.summary()
```

```python
plt.plot(df_train['평균기온'])
plt.plot(arima.predict())
plt.show()
```

```python
forecast = arima.forecast(steps=365)

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(10, 5)
fig.set_dpi(300)
ax.plot(df_train.index[-100:], df_train['평균기온'][-100:], color=COLORS['green'])
ax.plot(forecast, color=COLORS['red'])
plt.show()
```
