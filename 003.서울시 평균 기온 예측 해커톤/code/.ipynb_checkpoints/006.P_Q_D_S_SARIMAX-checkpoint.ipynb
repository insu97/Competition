# 정상성 검사

KPSS(Kwiatkowski-Phillips-Schmidt-Shin) 테스트와 ADF( Augmented Dickey-Fuller) 테스트는 시계열 데이터의 정상성 여부를 평가하기 위한 두 가지 주요 테스트입니다. 그러나 이 두 테스트는 서로 다른 방식으로 정상성을 평가하고 다른 가설을 검정합니다.

1. **ADF 테스트 (Augmented Dickey-Fuller Test):**
   - **목적:** ADF 테스트는 주어진 시계열 데이터가 단위근(unit root) 문제를 가지고 있는지를 검정하여 정상성을 평가합니다.
   - **귀무가설:** 주어진 시계열 데이터에 단위근이 존재한다는 가설을 설정합니다. 즉, 시계열 데이터가 정상성을 가지고 있지 않다는 것을 의미합니다.
   - **대립가설:** 시계열 데이터가 정상성을 가지고 있다는 가설입니다.
   - **검정 통계량:** ADF 통계량은 단위근이 존재하는 정도를 나타내며, 이 값이 유의수준보다 작을 경우 귀무가설을 기각하고 정상성을 가지고 있다고 판단합니다.

2. **KPSS 테스트 (Kwiatkowski-Phillips-Schmidt-Shin Test):**
   - **목적:** KPSS 테스트는 시계열 데이터가 정상성을 가지고 있는지를 평가합니다.
   - **귀무가설:** 주어진 시계열 데이터가 정상성을 가지고 있다는 가설을 설정합니다.
   - **대립가설:** 시계열 데이터에 추세나 구조적인 변동이 존재하여 정상성을 가지고 있지 않다는 가설입니다.
   - **검정 통계량:** KPSS 통계량은 시계열 데이터에 추세가 있는 정도를 나타내며, 이 값이 유의수준보다 크면 귀무가설을 기각하고 정상성을 가지고 있지 않다고 판단합니다.

요약하면, ADF 테스트는 시계열 데이터가 추세를 포함한 비정상 상태인지를 평가하고, KPSS 테스트는 시계열 데이터가 추세나 구조적 변동 없이 정상 상태인지를 평가합니다. 일반적으로 두 테스트를 결합하여 사용하여 시계열 데이터의 정상성을 더 정확하게 평가할 수 있습니다.

## kpss

```python
from statsmodels.tsa.stattools import kpss

def kpss_test(series, **kw):    
    stats, p_value, nlags, critical_values = kpss(series, **kw)
    
    print(f'KPSS Stat: {stats:.5f}')
    print(f'p-value: {p_value:.2f}')
    print(f'Lags: {nlags}')
    
    print(f'검증결과: {"비정상(non-stationary)" if p_value <= 0.05 else "정상(stationary)"} 시계열 데이터입니다.')
```

```python
kpss_test(df_train['평균기온'], nlags=30) # 정상이면 차분 필요없음
```

## adfuller

```python
from statsmodels.tsa.stattools import adfuller #ADF 검정
```

```python
def adfuller_test(series, **kw):    
    adf, p_value, nlags, number_of_observations, critical_values, _ = adfuller(series, **kw)
    
    print(f'ADF: {adf:.5f}')
    print(f'p-value: {p_value:.2f}')
    print(f'Lags: {nlags}')
    print(f'Number of Observations: {number_of_observations}')
    
    print(f'검증결과: {"비정상(non-stationary)" if p_value > 0.05 else "정상(stationary)"} 시계열 데이터입니다.')
```

```python
adfuller_test(df_train['평균기온'], maxlag=40)
```

```python
import statsmodels.api as sm
```

# 학습

```python
train_size = int(len(df_train) * 0.8)
train_data = df_train['평균기온'][:train_size]
test_data = df_train['평균기온'][train_size:]
```

```python
import itertools
```

```python
P = D = Q = range(0, 2)  # 예제로 0에서 1까지의 값을 사용
S = [12]  # 계절 주기 (예: 12는 1년 주기)
seasonal_pdq = list(itertools.product(P, D, Q, S))
```

```python
df_train.index.freq = 'D'
```

```python
# 최적 모델 찾기
best_aic = np.inf
best_seasonal_order = None

for param in seasonal_pdq:
    try:
        model = SARIMAX(df_train['평균기온'], order=(1, 0, 0), seasonal_order=param)
        results = model.fit(disp=False)
        aic = results.aic
        if aic < best_aic:
            best_aic = aic
            best_seasonal_order = param
    except:
        continue

print(f"Best Seasonal Order: {best_seasonal_order} with AIC: {best_aic}")
```

```python
# 모델 생성 및 학습
model = SARIMAX(df_train['평균기온'], order=(1, 0, 0),
                seasonal_order=best_seasonal_order)  # p,d,q,s
result = model.fit()
```

```python
plt.plot(df_train['평균기온'])
plt.plot(result.predict())
plt.show()
```
